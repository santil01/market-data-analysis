```python
import requests, bs4, re
import pandas as pd
result = []
def parser():
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:61.0) Gecko/20100101 Firefox/61.0'}
    w = requests.get('https://www.investing.com/equities/russia', headers=headers).text
    w = bs4.BeautifulSoup(w, 'html.parser') 
    # Находим таблицу значений
    w = w.find('table', {'id': 'cross_rate_markets_stocks_1'}).find('tbody')
    # Находим все акции
    items = w.find_all('tr')
    # Вытаскиваем для каждой акции её данные
    for item in range(len(items)):
        prices = items[item].find_all('td')
        prices = list(map(lambda k: k.next_element, prices))
        # Берем нужную нам информацию
        prices_dict = {'ticker':prices[1].next_element, 'close':prices[2],
                       'high':prices[3], 'low':prices[4], 'chg':prices[5], 'chg%':prices[6], 'vol':prices[7]}
        # Убираем запятые из значений
        prices_dict = dict(map(lambda k: re.sub(',','', str(k)), x) for x in prices_dict.items())
        # Сохраняем все полученные данные в список
        result.append(prices_dict)
    return result
parser()
df = pd.DataFrame(result, columns = ['ticker','close','high','low', 'chg', 'chg%', 'vol'])
df


```
